```python
from PIL import Image
import matplotlib.pyplot as plt

from diffae import DiffusionAutoEncodersInterface
```

```python
args = {
    'output': '../output/celebahq128',
    'model_ckpt': 'last_ckpt.pth',
    'clf_ckpt': 'clf_last_ckpt.pth',
}
mode = 'infer'
diffae = DiffusionAutoEncodersInterface(args, mode)
```

# Smiling

```python
image = Image.open('../assets/celebahq_sample/53.jpg')
plt.imshow(image)
```

```python
target_id = 31  # Smiling
s = [0.2, 0.4, 0.6, 0.8]
results = diffae.infer_manipulate(image, target_id=target_id, s=s)

fig, ax = plt.subplots(1, len(s), tight_layout=True)
for i, r in enumerate(results):
    ax[i].imshow(r['output'])
    ax[i].set_title(s[i])
    ax[i].axis('off')
fig.show()
```
